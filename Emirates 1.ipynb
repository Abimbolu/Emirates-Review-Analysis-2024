```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Define the base URL
base_url = "https://www.trustpilot.com/review/www.emirates.com"

# Initialize a list to store data
all_data = []

# Scrape reviews from the first 30 pages
for page in range(1, 2):  # Adjust the range as needed
    print(f"Scraping page {page}...")
    page_url = f"{base_url}?page={page}"
    driver.get(page_url)
    time.sleep(5)  # Allow time for the page to load

    # Scroll to the bottom to load reviews
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)

    # Locate review cards on the page
    review_cards = driver.find_elements(By.CLASS_NAME, "styles_reviewCard__hcAvl")
    print(f"Number of review cards found on page {page}: {len(review_cards)}")

# Extract data from each card
for card in review_cards:
    try:
        # Reviewer Name
        reviewer_name = card.find_element(By.CLASS_NAME, "typography_heading-xxs__QKBS8").text
    except:
        reviewer_name = "No name listed"

    try:
        # Reviewer Country
        country_and_reviews = card.find_elements(By.CLASS_NAME, "typography_body-m__xgxZ_")
        reviewer_country = country_and_reviews[0].text if len(country_and_reviews) > 0 else "No country listed"
    except:
        reviewer_country = "No country listed"

    try:
        # Number of Reviews
        number_of_reviews = country_and_reviews[1].text if len(country_and_reviews) > 1 else "No reviews count"
    except:
        number_of_reviews = "No reviews count"
        
    try:
        
        review_text = card.find_element(By.CLASS_NAME, "styles_reviewContent__0Q2Tg").text
    except:
        review_text = "No review text"
    try:
        star_rating = card.find_element(By.CSS_SELECTOR, "img").get_attribute("alt")
    except:
        star_rating = "No rating"

    all_data.append({
        "Number of Reviews": reviewer_country,
        "Reviewer Name": reviewer_name,
        "Reviwer Country": number_of_reviews,
        "Review_text": review_text,
        "Star Rating": star_rating
    })

# Convert the data into a DataFrame
df = pd.DataFrame(all_data)
print(df)

# Save the DataFrame to a CSV file
df.to_csv('emirates_reviews.csv', index=False, encoding='utf-8')

# Close the driver
driver.quit()
```
